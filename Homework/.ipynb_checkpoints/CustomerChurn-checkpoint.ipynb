# Assig-2 Predicting Customer Churn in Telecommunications

```python
# importing necessary library and algorithms
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
# reading dataframe
df = pd.read_csv(r"C:\Users\Admin\SCHAND_CODEUNNATI\ML\Dataset\customer.csv")
df
```

# Data Analysis and data preprocessing

```python
df.info()
```

```python
LE = LabelEncoder()
df["churn"] = LE.fit_transform(df["churn"])
df["ebill"] = LE.fit_transform(df["ebill"])
df["region"] = LE.fit_transform(df["region"])
df["payment_method"] = LE.fit_transform(df["payment_method"])
df
```

```python
cols = df[['age','region','tenure','phones','contracts','internet','ebill','payment_method']]
cols
```

# Data visualization

```python
for x in cols:
    plt.xlabel(x)
    plt.ylabel('churn')
    plt.scatter(cols[x],df['churn'])
    plt.show()
```

```python
# Standarlize
relevant_cols = df[["age","tenure"]]
stdscl = StandardScaler()
x = stdscl.fit_transform(relevant_cols)
y = df["churn"]
```

```python
# Cross validation
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

# Training logistic model

```python
lgclf = LogisticRegression()
lgclf.fit(x_train,y_train)
y_pred = lgclf.predict(x_test)
y_pred
```

### Classification report

```python
display(classification_report(y_test,y_pred))
lgclf.score(x_test,y_test)
```
