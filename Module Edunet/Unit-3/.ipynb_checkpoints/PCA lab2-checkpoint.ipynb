```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
%matplotlib inline
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
cancer_data = load_breast_cancer()
```

```python
print(type(cancer_data))
```

```python
cancer_data.keys()
```

```python
no_features=len(cancer_data['feature_names'])
```

```python
print(no_features)
```

```python
no_records=len(cancer_data['data'])
```

```python
print(no_records)
```

```python
df = pd.DataFrame(cancer_data['data'],columns=cancer_data['feature_names'])
```

```python
df
```

```python
df.head()
```

```python
scaler=StandardScaler()
```

```python
df_scaled=scaler.fit_transform(df)
```

```python
df_scaled
```

```python
pca=PCA(n_components=2)
```

```python
pc12 = pca.fit_transform(df_scaled)
```

```python
pc12.shape
```

```python
df_pca = pd.DataFrame(pc12,columns=['PC1','PC2'])
df_pca.head()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(
    x=df_pca['PC1'],
    y=df_pca['PC2'],
    c=cancer_data['target'],
    cmap='Set1')
plt.xlabel('First Principal Component - PC1')
plt.ylabel('Second Principal Component -PC2');
```

```python
x = df_pca
y = cancer_data['target']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
logR = LogisticRegression()
logR.fit(x_train, y_train)
predictions = logR.predict(x_test)
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
print(classification_report(y_test, predictions))
```

```python
inertia = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(range(2, 11), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=0)
cluster_assignments = kmeans.fit_predict(df_scaled)
df['Cluster'] = cluster_assignments
```

```python
plt.scatter(df_pca['PC1'], df_pca['PC2'], c=df['Cluster'], cmap='viridis')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Clustered Data')
plt.show()
```
